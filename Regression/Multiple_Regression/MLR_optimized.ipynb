```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset.head(15)
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

```python
print(pd.DataFrame(X).head(15))
print(pd.DataFrame(y).head(15))
```

```python
from sklearn.preprocessing import  OneHotEncoder
onehotencoder = OneHotEncoder(categories='auto') #Assigning 0,1,2 to countries is not a good idea since making numerical comparisons for countries is senseless
X = np.append(onehotencoder.fit_transform(X[:,3].reshape(-1,1)).toarray() , X[:,:-1] , axis = 1).astype(np.float64)
y = y.astype(np.float64)
X = X[:,1:]
print(pd.DataFrame(X).head(15))
print(pd.DataFrame(y).head(15))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train , y_train)


#Predicting the model on test set
y_predict = regressor.predict(X_test)
```

```python
import statsmodels.formula.api as sm
X = np.append(arr = np.ones((50,1)).astype(int) , values = X , axis = 1)
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
numVars = 6
sl=0.05
for i in range(0, numVars):
    regressor_OLS = sm.OLS(y, X_opt).fit()
    maxVar = max(regressor_OLS.pvalues).astype(float)
    if maxVar > sl:
        for j in range(0, numVars - i):
            if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                   X_opt = np.delete(X_opt, j, 1)
```

```python
print(pd.DataFrame(X_opt).head(15))
#The optimized set X
```

```python

from sklearn.model_selection import train_test_split
X_train_opt, X_test_opt, y_train, y_test =  train_test_split(X_opt,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor_opt = LinearRegression()
X_train_opt, X_test_opt, y_train, y_test =  train_test_split(X_opt,y,test_size = 0.2, random_state = 0)
regressor_opt.fit(X_train_opt , y_train)
y_pred_opt = regressor_opt.predict(X_test_opt) 
print("Actual data")
print(pd.DataFrame(y_test))
print("Output from non optimized model")
print(pd.DataFrame(y_predict))
print("Output from optimized model")
print(pd.DataFrame(y_pred_opt))
```
