Based on the DataSet lets evaluate what a person is likely to buy if he has purchased eggs

```python
DataSet = np.matrix(pd.read_csv("Market_Basket_Optimisation.csv" , header = None))
```

```python
Count = {}
a,b = DataSet.shape
min_support = 0.05
```

```python
for i in range(a):
    for j in range(b):
        try:
                Count[DataSet[i,j]] +=1
        except:
                Count[DataSet[i,j]] = 1
                
Count
```

```python
Support = {}

for i in Count:
    Support[i] = Count[i]/a
    if Support[i] < min_support:
        del Support[i]
del Support[np.nan]


Support
```

Items with min support of 0.05 are selected

```python
del Support["eggs"]
Eggs = []
for i in range(a):
    for j in range(b):
        if DataSet[i,j] == "eggs":
            Eggs.append(np.array(DataSet[i]))
            break
Confidence = {}

for i in Support:
    for j in range(20):
        if i in Eggs[j]:
            try:
                Confidence[i]+=1
            except:
                Confidence[i] = 1
                
for i in Confidence:
    Confidence[i] = Confidence[i]/len(Eggs)

    
Confidence
```

Confidence of Eggs with selected items (exceeding minimum support) 

```python
Lift = {}
for i in Confidence:
    Lift[i] = Support[i]/Confidence[i]
Lift
```

```python
import operator
sorted_d = sorted(Lift.items(), key=operator.itemgetter(1))
A = []
for i in sorted_d[::-1]:
    A.append(i[0])
k = 1
for i in A:
    print(str(k) + ". " + i)
    k+=1
```

Items listed in decreasing liklihood of being purchased by a person who has also purchased eggs
