```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset.head(20)
```

```python
X = dataset.iloc[:, 2:4].values.astype(np.float64)
y = dataset.iloc[:, 4].values.astype(np.float64)

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size = 0.25, random_state = 54)

# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

The Data is now pre-processed

Buiding KNN Model

```python
def Dist(A,B):
    Sum=0
    for i in range(len(A)):
        Sum+=(A[i] - B[i])**2
    return Sum**(1/2)

def KNN(X_train , Y_train , point , k):
    import operator
    distn = []
    
    for i in range(len(X_train[:,0])):
        distn.append((Dist(X_train[i,:],point) , i))
        distn.sort(key = operator.itemgetter(0))
    
    Sum=0
    for i in range(k):
        Sum+=Y_train[distn[i][1]]
    if Sum/k > 0.5:
        return 1
    else:
        return 0
        

    
def predict(X_train , y_train , X_test,k):
    y_predict = np.zeros((len(X_test[:,0]),1))
    for i in range(len(X_test[:,0])):
        
        y_predict[i,0] = KNN(X_train , y_train , X_test[i,:],k)
    return y_predict
```

```python
y_pred = predict(X_train , y_train , X_test,5)
pd.DataFrame(y_pred)
```

```python
pd.DataFrame(y_test)
```

Assesing Model

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

Number of correct predictions = sum of left diagonal = 93

Number of incorrect predictions = sum of right diagonal = 7
