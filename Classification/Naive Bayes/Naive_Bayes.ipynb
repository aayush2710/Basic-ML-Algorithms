```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import collections
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, 2:4].values.astype(np.float64)
y = dataset.iloc[:, 4].values.astype(np.float64)
dataset
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size = 0.25, random_state = 54)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
def classifier(X_train , y_train, X ,r):
    def func(x,y):
        return (x - X[0])**2 + (y-X[1])**2 - r**2
    L=[]
    for i in range(len(X_train[:,0])):
        x = X_train[i,0]
        y = X_train[i,1]
        
        if func(x,y)<=0:
            L.append([x,y,y_train[i]])
    y1 = 0
    y0=0        
    for i in L:
        if i[2] == 1:
            y1+=1
        else:
            y0+=1
    
    Px = len(L)/len(X_train[:,0])
    P1 =  collections.Counter(y_train)[1]/ len(y_train)
    P0 =  collections.Counter(y_train)[0]/ len(y_train)    
    Px1 = y1 / collections.Counter(y_train)[1]
    Px0 = y0 / collections.Counter(y_train)[0]
    
    p1x = Px1*P1/Px
    p0x = Px0*P0/Px
    if p1x>=p0x:
        return 1
    else:
        return 0

def predict(X_train,y_train,X_test,r):
    y_predict = np.ones(len(X_test[:,0]))
    for i in range(len(X_test[:,0])):
        X = X_test[i,:]
        if classifier(X_train, y_train , X,r)==0:
            y_predict[i] = 0
    return y_predict
```

```python
y_pred = predict(X_train,y_train,X_test,0.5)
pd.DataFrame(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

No of incorrect predictions = Sum of Right Diagonal = 7
